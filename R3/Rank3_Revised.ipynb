```python
import pandas as pd
import plotly.graph_objects as go
```

```python
zip_df = pd.read_csv(r"Z:\ArcGIS Pro\Fabiah\Bottleneck\zip codes\R3\top15_od_pairs_3.csv")
county_df = pd.read_csv(r"Z:\ArcGIS Pro\Fabiah\Bottleneck\County FULL 2025\R3\top15_od_pairs_counties_3.csv")
```

```python
import pandas as pd
import plotly.graph_objects as go

def make_sankey(df, title):
    bottleneck_title = "2025 Bottleneck Rank 3: I-495 CCW @ MD-97/GEORGIA AVE/EXIT 31"
    max_effect_periods = "Weekday-AM Peak"

    plot_df = df.copy()

    # Clean fields
    plot_df["origin"] = plot_df["origin"].astype(str).str.strip()
    plot_df["destination"] = plot_df["destination"].astype(str).str.strip()
    plot_df["volume"] = pd.to_numeric(plot_df["volume"], errors="coerce")

    plot_df = plot_df.dropna(subset=["origin", "destination", "volume"])
    plot_df = plot_df[plot_df["volume"] > 0]
    plot_df = plot_df[plot_df["origin"] != plot_df["destination"]]

    # Rank OD pairs by volume
    plot_df = plot_df.sort_values("volume", ascending=False).reset_index(drop=True)
    plot_df["od_rank"] = plot_df.index + 1
    plot_df["share"] = plot_df["volume"] / plot_df["volume"].sum()
    plot_df["od_pair"] = plot_df["origin"] + " → " + plot_df["destination"]

    # ------------------------------------------------------------------
    # shared-node Sankey structure
    # ------------------------------------------------------------------
    labels = pd.unique(
        pd.concat(
            [plot_df["origin"], plot_df["destination"]],
            ignore_index=True
        )
    )

    label_to_index = {
        label: i for i, label in enumerate(labels)
    }

    sources = plot_df["origin"].map(label_to_index)
    targets = plot_df["destination"].map(label_to_index)
    values = plot_df["volume"]

    customdata = plot_df[[
        "od_rank",
        "origin",
        "destination",
        "share"
    ]].values

    fig = go.Figure(data=[go.Sankey(
        arrangement="snap",
        domain=dict(x=[0.04, 0.94], y=[0.30, 0.88]),
        node=dict(
            pad=24,
            thickness=22,
            line=dict(color="black", width=0.5),
            label=labels.tolist(),
            hovertemplate=(
                "<b>%{label}</b>"
                "<extra></extra>"
            )
        ),
        link=dict(
            source=sources,
            target=targets,
            value=values,
            customdata=customdata,
            hovertemplate=(
                "<b>OD Pair Rank %{customdata[0]}</b><br>"
                "%{customdata[1]} → %{customdata[2]}<br>"
                "Share of shown top-15 trips: %{customdata[3]:.1%}"
                "<extra></extra>"
            )
        )
    )])

    # ------------------------------------------------------------------
    # Summary statistics for key finding
    # ------------------------------------------------------------------
    total_volume = plot_df["volume"].sum()
    top_pair = plot_df.iloc[0]
    top_5_share = plot_df.head(5)["volume"].sum() / total_volume

    dominant_origin = (
        plot_df.groupby("origin")["volume"]
        .sum()
        .sort_values(ascending=False)
    )

    dominant_destination = (
        plot_df.groupby("destination")["volume"]
        .sum()
        .sort_values(ascending=False)
    )

    top_origin = dominant_origin.index[0]
    top_origin_share = dominant_origin.iloc[0] / total_volume

    top_destination = dominant_destination.index[0]
    top_destination_share = dominant_destination.iloc[0] / total_volume

    finding_text = (
        f"<b>Key finding:</b> During the maximum-effect period(s), "
        f"<b>{max_effect_periods}</b>, the shown top-15 OD flows are mainly driven by "
        f"<b>{top_origin}</b> as the leading origin "
        f"({top_origin_share:.1%} of shown trips) and "
        f"<b>{top_destination}</b> as the leading destination "
        f"({top_destination_share:.1%}).<br>"
        f"The largest OD pair is <b>{top_pair['origin']} → {top_pair['destination']}</b> "
        f"({top_pair['share']:.1%} of shown trips), and the top 5 OD pairs account for "
        f"<b>{top_5_share:.1%}</b> of shown trips."
    )

    # ------------------------------------------------------------------
    # Ranked table below Sankey — volume removed, share retained
    # ------------------------------------------------------------------
    table_df = plot_df.head(15).copy()
    table_df["share_text"] = table_df["share"].map(lambda x: f"{x:.1%}")

    fig.add_trace(
        go.Table(
            domain=dict(x=[0.04, 0.94], y=[0.0, 0.24]),
            header=dict(
                values=[
                    "<b>OD Rank</b>",
                    "<b>Origin</b>",
                    "<b>Destination</b>",
                    "<b>Share of Shown Trips</b>"
                ],
                fill_color="#F1F3F5",
                align=["center", "left", "left", "right"],
                font=dict(size=14, color="#24313A"),
                height=28
            ),
            cells=dict(
                values=[
                    table_df["od_rank"],
                    table_df["origin"],
                    table_df["destination"],
                    table_df["share_text"]
                ],
                fill_color="white",
                align=["center", "left", "left", "right"],
                font=dict(size=12, color="#24313A"),
                height=25
            )
        )
    )

    fig.update_layout(
        title=dict(
            text=(
                f"<b>{title}</b><br>"
                f"<span style='font-size:22px'><b>{bottleneck_title}</b></span><br>"
                f"<span style='font-size:16px'>Maximum-effect period(s): "
                f"{max_effect_periods}</span>"
            ),
            x=0.02,
            y=0.98,
            xanchor="left",
            yanchor="top"
        ),
        font=dict(size=14, family="Arial", color="#050608"),
        width=1250,
        height=1040,
        paper_bgcolor="white",
        plot_bgcolor="white",
        margin=dict(l=60, r=110, t=250, b=30),
        annotations=[
            dict(
                text=finding_text,
                x=0.04,
                y=1.055,
                xref="paper",
                yref="paper",
                showarrow=False,
                align="left",
                font=dict(size=14, color="#24313A"),
                bgcolor="rgba(245,245,245,0.98)",
                bordercolor="#D8DEE3",
                borderwidth=1,
                borderpad=10
            ),
            dict(
                text="<b>Flow direction: Origin → Destination</b>",
                x=0.04,
                y=0.94,
                xref="paper",
                yref="paper",
                showarrow=False,
                align="left",
                font=dict(size=14, color="#4A555C")
            )
        ]
    )

    return fig
```

```python
county_sankey = make_sankey(county_df, "Top 15 Jurisdiction O-D Flows")
county_sankey
```

```python
county_sankey.write_html("county_od_sankey.html")
```

```python
zip_sankey = make_sankey(zip_df, "Top 15 ZIP Code O-D Flows")
zip_sankey.write_html("zip_od_sankey.html")
```
